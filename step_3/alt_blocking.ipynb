```python
import py_entitymatching as em
import pandas as pd
```

```python
source1 = '/Users/shantanusinghal/workspace/cs839DataScience/step_3/source1_cleaned.csv'
source2 = '/Users/shantanusinghal/workspace/cs839DataScience/step_3/source2_cleaned.csv'

# Read the data
A = em.read_csv_metadata(source1, key = 'ID')
B = em.read_csv_metadata(source2, key = 'ID')
```

```python
em.get_attr_types(A).keys()
```

```python
block_t = em.get_tokenizers_for_blocking()
block_s = em.get_sim_funs_for_blocking()
block_c = em.get_attr_corres(A, B)
atypes_A = em.get_attr_types(A)
atypes_B = em.get_attr_types(B)

block_f = em.get_features(A, B, atypes_A, atypes_B, block_c, block_t, block_s)
block_f
```

```python
#Blocking plan :

#A, B --rule-based-blocking--> candset(C0)-----|                                |-----black-box-blocking--> candset(C2)-----|  
#                                              |----union--->candset (C01) ---> |                                           |----union--->candset (C) 
#A, B --rule-based-blocking--> candset(C1)-----|                                |-----black-box-blocking--> candset(C3)-----|

```

```python
# Rule-Based blocker [C0]
rb0 = em.RuleBasedBlocker()

# add rule for book names : block tuples if Levenshtein Similarity is below 0.3
rb0.add_rule(['Name_Name_lev_sim(ltuple, rtuple) < 0.5'], block_f)

# add rule for authors : block tuples if Jaccard Similarity is below 0.3 in spaces delimited tokens
rb0.add_rule(['Author_Author_jac_dlm_dc0_dlm_dc0(ltuple, rtuple) < 0.5'], block_f)

C0 = rb0.block_tables(A, B,
                    l_output_attrs=['ID', 'Name', 'Author', 'Publisher', 'Publishing_Date', 'Format', 'Pages', 'Rating'], 
                    r_output_attrs=['ID', 'Name', 'Author', 'Publisher', 'Publishing_Date', 'Format', 'Pages', 'Rating'])
C0
```

```python
# Rule-Based blocker [C1]
rb1 = em.RuleBasedBlocker()

# add rule for book names : block tuples if Levenshtein Similarity is below 0.3
rb1.add_rule(['Name_Name_lev_sim(ltuple, rtuple) < 0.67'], block_f)

# add rule for publishers : block tuples if Jaccard Similarity is below 0.3 in three-gram tokens
rb1.add_rule(['Publisher_Publisher_jac_qgm_3_qgm_3(ltuple, rtuple) < 0.67'], block_f)

C1 = rb1.block_tables(A, B,
                    l_output_attrs=['ID', 'Name', 'Author', 'Publisher', 'Publishing_Date', 'Format', 'Pages', 'Rating'], 
                    r_output_attrs=['ID', 'Name', 'Author', 'Publisher', 'Publishing_Date', 'Format', 'Pages', 'Rating'])
C1
```

```python
# Combining Blocker Outputs of C0 and C1
C01 = em.combine_blocker_outputs_via_union([C0, C1], verbose=True)
C01
```

```python
# Blackbox Blockers [C2]
bb0 = em.BlackBoxBlocker()

# add rule for page count : block tuples if page count differs by more than 25
def range_20(ltuple, rtuple):
    return abs(ltuple['Pages'] - rtuple['Pages']) > 20

bb0.set_black_box_function(range_20)

C2 = bb0.block_candset(C01)
C2
```

```python
from datetime import datetime

def same_year(ltuple, rtuple):
    try:
        l_year = datetime.strptime(str(ltuple['Publishing_Date']), '%Y-%m-%d').year
        r_year = datetime.strptime(str(rtuple['Publishing_Date']), '%Y-%m-%d').year
        return l_year != r_year
    except:
        return True

# Blackbox Blockers
bb1 = em.BlackBoxBlocker()
bb1.set_black_box_function(same_year)

C3 = bb1.block_candset(C01)
C3
```

```python
# Combining Blocker Outputs of C2 and C3
C = em.combine_blocker_outputs_via_union([C2, C3], verbose=True)
C
```

```python
# Save to file
em.to_csv_metadata(C, './candidate_set.csv')
```
