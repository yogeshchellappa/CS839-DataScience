```python
# Import required libraries
import py_entitymatching as em
import os
import pandas as pd
import pandas_profiling
```

```python
source1 = 'source1_cleaned.csv'
source2 = 'source2_cleaned.csv'

# Read the data
data_source1 = em.read_csv_metadata(source1)
data_source2 = em.read_csv_metadata(source2)
```

```python
# Set the metadata
em.set_key(data_source1, 'ID')
em.set_key(data_source2, 'ID')
```

```python
# Check
print (em.get_key(data_source1))
print (em.get_key(data_source2))

print (data_source1.head())
print (data_source2.head())
```

```python
# Get data profile and save to file
profile_1 = pandas_profiling.ProfileReport(data_source1)
profile_2 = pandas_profiling.ProfileReport(data_source2)

profile_1.to_file('source_1_profile.html')
profile_2.to_file('source_2_profile.html')

print (profile_1)
print (profile_2)
```
